```python
# Looking at Target Column ('Diabetes') for distinct values

print(df['Diabetes'].value_counts())
```

```python
# Decode Diabetes Column

pd.set_option('future.no_silent_downcasting', True)
df['Diabetes'] = df['Diabetes'].replace({0: 'Non-Diabetic', 1: 'Pre-Diabetic', 2: 'Diabetic'})
df.head()
```

```python
# Decode High Blood Pressure Column with random values 110 - 220. 110 - 130 Normal and 131 - 220 high BP

pd.set_option('future.no_silent_downcasting', True)
df["HighBP"] = df["HighBP"].apply(lambda x: np.random.randint(110, 130) if x == 0 else np.random.randint(131, 220))
df.head()
```

```python
# Decode High Colestrol Column with random values 120 - 400. 120 - 200 Normal and 201 - 400 high Cholestrol

pd.set_option('future.no_silent_downcasting', True)
df["HighChol"] = df["HighChol"].apply(lambda x: np.random.randint(120, 201) if x == 0 else np.random.randint(201, 401))
df.head()
```

```python
# Decode HvyAlcoholConsump.. Normal 1 to 10 drinks 

df.rename(columns={'HvyAlcoholConsump': 'AlcoholConsump'}, inplace=True)
pd.set_option('future.no_silent_downcasting', True)
df["AlcoholConsump"] = df["AlcoholConsump"].apply(lambda x: np.random.randint(0, 10) if x == 0 else np.random.randint(11, 20))
df.head()
```

```python
# Rename column name- Have any kind of health care coverage, including health insurance, prepaid plans such as HMO, etc. 0 = no 1 = yes


df.rename(columns={'AnyHealthcare': 'HealthCoverage'}, inplace=True)
df.head()
```

```python
# Drop Column NoDocbcCost

df = df.drop('NoDocbcCost', axis=1)
df.head()
```

```python
# Decode Would you say that in general your health is: scale 1-5 1 = excellent 2 = very good 3 = good 4 = fair 5 = poor

pd.set_option('future.no_silent_downcasting', True)
df['GenHlth'] = df['GenHlth'].replace({5: 'Excellent', 4: 'Very Good', 3: 'Good', 2: 'Fair', 1: 'Poor'})
df.head()
```

```python
# Decode Do you have serious difficulty walking or climbing stairs? 0 = no 1 = yes

pd.set_option('future.no_silent_downcasting', True)
df["DiffWalk"] = df["DiffWalk"].replace({0:'No Difficulty', 1:'Face Difficulty'})
df.head()
```

```python
# Decode Do you have serious difficulty walking or climbing stairs? 0 = no 1 = yes

df.rename(columns={'Sex': 'Gender'}, inplace=True)
pd.set_option('future.no_silent_downcasting', True)
df["Gender"] = df["Gender"].replace({0:'Female', 1:'Male'})
df.head()
```

```python
# Decode Age column by randomly assigning any value within the ranges defined in the dataset description

pd.set_option('future.no_silent_downcasting', True)

# Function to generate random age based on the age category 
def generate_random_age(age_category): 
    if age_category == 1: 
        return random.randint(18, 24) 
    elif age_category == 2: 
        return random.randint(25, 29) 
    elif age_category == 3: 
        return random.randint(30, 34) 
    elif age_category == 4: 
        return random.randint(35, 39) 
    elif age_category == 5: 
        return random.randint(40, 44) 
    elif age_category == 6: 
        return random.randint(45, 49) 
    elif age_category == 7: 
        return random.randint(50, 54) 
    elif age_category == 8: 
        return random.randint(55, 59) 
    elif age_category == 9: 
        return random.randint(60, 64) 
    elif age_category == 10: 
        return random.randint(65, 69) 
    elif age_category == 11: 
        return random.randint(70, 74)
    elif age_category == 12: 
        return random.randint(75, 79)
    elif age_category == 13: 
        return random.randint(80, 98)
    else: return None 

# Apply the function to the DataFrame column 
df['Age'] = df['Age'].apply(generate_random_age)
df.head()
```

```python
# Decode Education column by randomly assigning any value within the ranges defined in the dataset description

pd.set_option('future.no_silent_downcasting', True)

# Function to generate random age based on the age category 
def generate_Education(edu_category): 
    if edu_category == 1: 
        return "Never attended School" 
    elif edu_category == 2: 
        return "Elementary Education" 
    elif edu_category == 3: 
        return "High School" 
    elif edu_category == 4: 
        return "High School Graduate" 
    elif edu_category == 5: 
        return "Technical College" 
    elif edu_category == 6: 
        return "College Graduate" 
    else: return None 

# Apply the function to the DataFrame column 
df['Education'] = df['Education'].apply(generate_Education)
df.head()
```

```python
# Decode Income column by randomly assigning any value within the ranges defined in the dataset description

pd.set_option('future.no_silent_downcasting', True)

# Function to generate random age based on the age category 
def generate_random_income(income_category): 
    if income_category == 1: 
        return random.randint(5000, 9999) 
    elif income_category == 2: 
        return random.randint(10000, 15999) 
    elif income_category == 3: 
        return random.randint(16000, 21999) 
    elif income_category == 4: 
        return random.randint(22000, 27999) 
    elif income_category == 5: 
        return random.randint(28000, 34999) 
    elif income_category == 6: 
        return random.randint(35000, 54999) 
    elif income_category == 7: 
        return random.randint(55000, 74999) 
    elif income_category == 8: 
        return random.randint(75000, 100000) 
    else: return None 

# Apply the function to the DataFrame column 
df['Income'] = df['Income'].apply(generate_random_income)
df.head()
```
